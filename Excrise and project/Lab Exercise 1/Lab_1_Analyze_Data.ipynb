```python
import pandas as pd

excel_file_path = "Sample - Superstore(1).xls"

df = pd.read_excel(excel_file_path)
# Display the contents of the DataFrame
df.head()
```

```python
# Identify the completeness of each column
completeness = df.count() / len(df) * 100
completeness
```

```python
# Identify inconsistencies in each column
inconsistencies = {}
for column in df.columns:
    unique_values = df[column].unique()
    if len(unique_values) > 1:
        inconsistencies[column] = unique_values

inconsistencies
```

```python
# Identify duplicate rows
duplicate_rows = df[df.duplicated(keep=False)]
duplicate_rows.sum()
```

```python
# Summarize the data
print("Completeness Summary:")
print(completeness)
print("\nInconsistencies Summary:")
print(inconsistencies)
print("\nDuplicate Rows:")
print(duplicate_rows)
print("\nData Summary:")
print(df.describe())
```
